```python
import pandas as pd
import numpy as np
import os
```

```python
inpath = r'stop_words/'
outpath = r'stop_words/SWs.txt'
outpath2 = r'stop_words/imageWords.txt'
```

```python
# fileList = os.listdir(inpath)
fileList = ['character_entity.csv',
 'concept_entity.csv',
 'event_entity.csv',
 'goods_entity.csv',
 'health_entity.csv',
 'medical_entity.csv',
 'prevention_entity.csv',
 'wiki_entity.csv']
```

```python
stop_words = []
for name in fileList:
    cache = pd.read_csv(inpath + name, encoding='utf-8')
    cache = cache.drop(cache[cache[':LABEL'].isin(['事迹','事件','言论'])].index)
    stop_words = stop_words + list(cache.drop(cache[cache[':LABEL'].str.contains('String|Date')].index)['value'].str.replace('《|》','').str.strip())
stop_words = set(stop_words)
```
